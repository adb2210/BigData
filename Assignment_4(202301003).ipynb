```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('/content/company_sales_data.csv')
plt.plot(data['month_number'])
plt.plot(data['total_profit'])
plt.xlabel('month')
plt.ylabel('profit')
plt.show()
```

```python
plt.plot(data['facecream'],label='facecream')
plt.plot(data['facewash'],label='facewash')
plt.plot(data['toothpaste'],label='toothpaste')
plt.plot(data['bathingsoap'],label='bathingsoap')
plt.plot(data['shampoo'],label='shampoo')
plt.plot(data['moisturizer'],label='moisturizer')
plt.xlabel('month')
plt.ylabel('total Sales')
plt.legend()
plt.show()
```

```python
plt.bar(1,max(data['facecream']),label='facecream')
plt.bar(2,max(data['facewash']),label='facewash')
plt.bar(3,max(data['toothpaste']),label='toothpaste')
plt.bar(4,max(data['bathingsoap']),label='bathingsoap')
plt.bar(5,max(data['shampoo']),label='shampoo')
plt.bar(6,max(data['moisturizer']),label='moisturizer')
plt.xlabel('product')
plt.ylabel('max Sale')
plt.legend()
plt.show()
```

```python
plt.scatter(data['month_number'],data['facecream'],label='facecream')
plt.scatter(data['month_number'],data['facewash'],label='facewash')
plt.scatter(data['month_number'],data['toothpaste'],label='toothpaste')
plt.scatter(data['month_number'],data['bathingsoap'],label='bathingsoap')
plt.xlabel('month')
plt.ylabel('products')
plt.legend()
plt.show()
```
