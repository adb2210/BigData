```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([2,4,6,8])
y=np.array([3,7,5,10])
n=np.size(x)
total_x=np.sum(x)
total_y=np.sum(y)
x2=0
for i in x:
  x2=x2+i*i
xy=0
for i in range(n):
  xy=xy+x[i]*y[i]
b0=(total_y*x2-total_x*xy)/(n*x2-total_x*total_x)
b1=(n*xy-total_x*total_y)/(n*x2-total_x*total_x)
print('Value of b0:',b0)
print('Value of b1:',b1)
inputX=int(input('Enter value of x:'))
newY=b0+b1*inputX
print('Value of y:',newY)
x=np.append(x,inputX)
y=np.append(y,newY)
plt.scatter(x,y,color='red')
plt.plot(x,b0+b1*x,color='green')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
